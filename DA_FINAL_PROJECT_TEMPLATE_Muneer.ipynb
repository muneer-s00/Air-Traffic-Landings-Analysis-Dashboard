## Stage 3 – EDA and Visualizations

**1. Histogram – Landing Count Distribution**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["Landing Count"])
plt.title("Distribution of Landing Count")
plt.xlabel("Landing Count")
plt.ylabel("Frequency")
plt.show()
```

**What the chart shows**

A histogram displaying how frequently different landing count values occur.

**Features used**

Landing_Count

**What it is saying**

Most landing count values lie in the lower range, and very high landing counts are rare.

**Insight / What we are showing**

This reveals a right-skewed distribution, indicating that typical landings are small, and only a few records have extremely high activity.
This helps understand the dataset’s natural distribution and plan statistical methods accordingly.

**2. Histogram – Total Landed Weight**

```python
plt.figure(figsize=(8,5))
plt.hist(df["Total Landed Weight"])
plt.title("Distribution of Total Landed Weight")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.show()
```

**What the chart shows**

Frequency distribution of total landed weight.

**Features used**

Total_Landed_Weight

**What it is saying**

Most weights are small or moderate; extremely large weights are very few.

**Insight / What we are showing**

Also shows right-skewed distribution, which is typical in fishing/landing datasets.
It indicates that few species/operations dominate the total weight, important for cost forecasting and operational planning.

**3. Countplot – Species Frequency**

```python
plt.figure(figsize=(10,6))
df["Operating Airline"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Most Active Airlines")
plt.xlabel("Airline")
plt.ylabel("Count of Landings")
plt.xticks(rotation=45)
plt.show()
```

**What the chart shows**

The most frequently landed species.

**Features used**

Species_Name (count)

**What it is saying**

A few species dominate the overall landing operations.

**Insight / What we are showing**

Helps identify which species are most commonly caught, enabling:

* trend analysis

* supply planning

* price/market predictions.


**4. Countplot – Year Distribution**

```python
df['Year'] = df['Activity Period'].astype(str).str[:4].astype(int)
plt.figure(figsize=(8,5))
df["Year"].value_counts().sort_index().plot(kind="bar")
plt.title("Data Records Per Year")
plt.xlabel("Year")
plt.ylabel("Count of Records")
plt.show()
```

**What the chart shows**

Annual distribution of the number of records.

**Features used**

Year (count)

**What it is saying**

Shows if the dataset is balanced or biased toward certain years.

**Insight / What we are showing**

This helps determine:

* data completeness across years

* whether a specific year had unusually high or low reporting

* whether time-series analysis will be stable or irregular.




**5. Countplot – Port-wise Activity**

```python
plt.figure(figsize=(10,5))
df["GEO Region"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Most Active Regions")
plt.xlabel("Region")
plt.ylabel("Number of Landings")
plt.xticks(rotation=45)
plt.show()
```

**What the chart shows**

Top 10 ports with highest landings.

**Features used**

Port_Code (count)

**What it is saying**

Certain ports have significantly higher landing activity.

**Insight / What we are showing**

Shows geographical dominance—useful for:

* allocating resources

* identifying revenue-generating ports

* focusing on high-volume regions.

**6. Scatter Plot – Landing Count vs Total Weight**

```python
plt.figure(figsize=(6,5))
plt.scatter(df["Landing Count"], df["Total Landed Weight"])
plt.title("Landing Count vs Total Landed Weight")
plt.xlabel("Landing Count")
plt.ylabel("Total Landed Weight")
plt.show()
```

**What the chart shows**

The relationship between landing count and total landed weight.

**Features used **

Landing_Count

Total_Landed_Weight

**What it is saying**

When landing count increases, total weight also increases — but not perfectly.

**Insight / What we are showing**

Indicates a moderate positive correlation.
Helps understand:

* whether landing activity level predicts total weight

* operational efficiency

* species composition effects.

**7. Boxplot – Weight Distribution by Species (Top 5)**

```python
top_airlines = df["Operating Airline"].value_counts().head(5).index
df_top_airlines = df[df["Operating Airline"].isin(top_airlines)]

plt.figure(figsize=(10,6))
df_top_airlines.boxplot(column="Total Landed Weight", by="Operating Airline")
plt.title("Weight Distribution by Top 5 Operating Airlines")
plt.suptitle("")
plt.xlabel("Operating Airline")
plt.ylabel("Total Landed Weight")
plt.xticks(rotation=45)
plt.show()
```

**What the chart shows**

Weight variation across the 5 most frequent species.

**Features used**

Species_Name

Total_Landed_Weight

**What it is saying**

Each species has a different weight profile (median, spread, outliers).

**Insight / What we are showing**

Helps identify:

* which species contribute highest weight

* variability in weight

* species-specific risk or seasonal effects.

Useful for business decisions related to pricing and logistics.

**8. Bar Plot – Monthly Trend of Total Weight**

```python
df['Month'] = df['Activity Period'].astype(str).str[4:6].astype(int)
monthly = df.groupby("Month")["Total Landed Weight"].sum()

plt.figure(figsize=(8,5))
monthly.plot(kind="bar")
plt.title("Monthly Total Landing Weight")
plt.xlabel("Month")
plt.ylabel("Total Weight")
plt.show()
```

**What the chart shows**

Total monthly landing weight aggregated across the dataset.

**Features used**

Month

Total_Landed_Weight (aggregated)

**What it is saying**

Some months have higher weights, showing seasonality.

**Insight / What we are showing**

Reveals seasonal trends:

* peak months for fishing

* months with lower activity

* possible effects of weather, migration, or regulations.

This is valuable for planning manpower, storage, and budget cycles.

**9. Correlation Heatmap – Numeric Features**

```python
plt.figure(figsize=(8,6))
plt.imshow(df[["Landing Count","Total Landed Weight"]].corr(), cmap="viridis")
plt.colorbar()
plt.xticks([0,1], ["Landing Count","Total Weight"])
plt.yticks([0,1], ["Landing Count","Total Weight"])
plt.title("Correlation Heatmap")
plt.show()
```

**What the chart shows**

Correlation between numerical variables.

**Features used**

Landing_Count

Total_Landed_Weight

**What it is saying**

Shows how strongly numeric features relate to one another.

**Insight / What we are showing**

This helps with:

* feature engineering

* identifying redundancies

* model building decisions

Example: A high correlation means one variable may predict another.
